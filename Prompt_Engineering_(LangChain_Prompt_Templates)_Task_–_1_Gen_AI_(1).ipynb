 **Task 1**

Convert hardcoded prompt into reusable PromptTemplate.

**Task 2**

Create a prompt with multiple inputs: topic, audience, tone.

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

**Task 3**

Create multiple templates for different styles: teaching, interview, storytelling.

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

**Task 4**

Build a chat-based prompt system using roles like teacher, interviewer, and motivator.

```python
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_msg = "You are an interviewer who asks insightful questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker who inspires learning."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Tell me about {topic}")
    ])

    return chat_template.format_messages(topic=topic)

print(get_chat_prompt("teacher", "Neural Networks"))
```

**Task 5**

Validate user inputs for audience and tone.

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience → defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → defaulting to 'formal'")
        tone = "formal"

    return audience, tone

print(validate_inputs("kids", "funny"))
```

**Task 6**

Generate prompts dynamically based on style.

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )
    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} tone"
        )
    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    else:
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    return template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7**

Test one template with multiple inputs to ensure reusability.

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "formal"),
    ("Python", "intermediate", "casual"),
    ("Data Science", "expert", "formal"),
    ("Machine Learning", "beginner", "fun"),
    ("Deep Learning", "expert", "technical")
]

for topic, audience, tone in inputs:
    audience, tone = validate_inputs(audience, tone)
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
```
